# Matplot圖表繪製功能示範

```
import numpy as np
from matplotlib import pyplot as plt
import random

x = np.arange(1, 15)
y1=[]
y2 = []

for xi in x:
    y1.append(2 * xi + (random.randint(1,20)))
print(y1)

for xi in x:
    y2.append(2 * xi + (random.randint(-10,20)))
print(y2)
plt.title("Matplotlib demo")
plt.xlabel("x caption")
plt.ylabel("y Data")

plt.plot(x, y1)
plt.plot(x, y2, ':r')
#df.plot(kind = 'line', y = ['匯率','現金'])
plt.show()
```
